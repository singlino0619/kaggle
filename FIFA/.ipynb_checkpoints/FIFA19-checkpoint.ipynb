```python
train_df['Value_num'] = train_df['Value'].str.extract('([0-9]+)').astype(int)
```

```python
train_df['Value_unit'] = train_df['Value'].str.extract('([a-zA-Z])')
```

```python
train_df['Value_unit'] = train_df['Value_unit'].fillna('K')
```

```python
train_df['Value_unit_num'] = train_df['Value_unit'].map({'M': 1000000, 'K': 1000}).astype(int)
```

```python
train_df['Value_jpy'] = train_df['Value_unit_num'] * train_df['Value_num'] * 120
```

```python
sns.jointplot(train_df['Overall'], train_df['Value_jpy'], train_df)
```

```python
# complete Preferred Foot with 'Right'.
train_df.loc[train_df['Preferred Foot'].isnull(), 'Preferred Foot'] = 'Right'
```

```python
# mapping character to number
train_df['Preferred Foot'] = train_df['Preferred Foot'].map({'Right': 1, 'Left': 2}).astype(int)
```

```python
# complete International reputation to 1 which is lowest number among them.
train_df.loc[train_df['International Reputation'].isnull(), 'International Reputation'] =1
```

```python
X_train = train_df[['Preferred Foot', 'Overall', 'International Reputation']][:-100]
X_test = train_df[['Preferred Foot', 'Overall', 'International Reputation']][-100:]
Y_train = train_df['Value_jpy'][:-100]
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_predict = logreg.predict(X_test)
acc_log = logreg.score(X_train, Y_train)
acc_log
```

```python
linreg = LinearRegression()
linreg.fit(X_train, Y_train)
Y_predict = linreg.predict(X_test)
acc_log = linreg.score(X_train, Y_train)
acc_log
```

```python
linreg.coef_
```
